# Genero el Feature Engineering solicitado para el DataFrame 'users_reviews'

Feature Engineering: En el dataset user_reviews se incluyen reseñas de juegos hechos por distintos usuarios. Debes crear la columna 'sentiment_analysis' aplicando análisis de sentimiento con NLP con la siguiente escala: debe tomar el valor '0' si es malo, '1' si es neutral y '2' si es positivo. Esta nueva columna debe reemplazar la de user_reviews.review para facilitar el trabajo de los modelos de machine learning y el análisis de datos. De no ser posible este análisis por estar ausente la reseña escrita, debe tomar el valor de 1.

```python
users_reviews
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk
```

```python
# Se instala una vez!
# nltk.download('vader_lexicon')
```

```python
# 'review' es la columna que quiero convertir a tipo str
users_reviews['review'] = users_reviews['review'].astype(str)

# Verifica si todos los elementos ahora son de tipo str
todos_son_str = users_reviews['review'].apply(lambda x: isinstance(x, str)).all()

# Imprimir el resultado
print("¿Todos los elementos son de tipo str?", todos_son_str)
```

```python
# Creeo una instancia del SentimentIntensityAnalyzer con la variable sia (inciales)
sia = SentimentIntensityAnalyzer()

# Agrego la columna solicitada y se le asigna los valores
users_reviews['sentiment_analysis'] = users_reviews['review'].apply(lambda x: sia.polarity_scores(x)['compound'])

# Asigna valores según la escala dada
users_reviews['sentiment_analysis'] = users_reviews['sentiment_analysis'].apply(lambda x: 0 if x < -0.05 else (2 if x > 0.05 else 1))

# En caso de reseñas ausentes, asigna el valor 1
users_reviews['sentiment_analysis'].fillna(1, inplace=True)

# Elimina la columna original de 'reviews' 
users_reviews.drop('review', axis=1, inplace=True)
```

```python
users_reviews.head(2)
```

```python
# En la variable counts guardo la cantidad de reseñas de cada tipo
counts = users_reviews['sentiment_analysis'].value_counts()

# Muestro la cantidad de datos malos, neutrales y positivos. Solamente para tener nocion de cantidades
print(counts)
```
